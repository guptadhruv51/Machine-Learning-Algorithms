### Missing Values

```python
feature_with_na=[features for features in train_data.columns if train_data[features].isna().sum() >1]
for i in feature_with_na:
    print(i,": ",np.round(train_data[i].isnull().mean(),4),' % missing values')
```

```python
for feature in feature_with_na:
    data=train_data.copy()
    data[feature]=np.where(data[feature].isna(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
len(np.where(train_data[feature_with_na[0]].isna(),1,0))
```

### Numerical Features

```python
numerical_features=[feature for feature in train_data.columns if train_data[feature].dtypes!='O']
print("Number of Numerical Features: ",len(numerical_features))
train_data[numerical_features].head()
```

```python
year_features=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
for feature in year_features:
    print(feature,train_data[feature].unique())
```

```python
train_data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
for feature in year_features:
    if feature!='YrSold':
        data=train_data.copy()
    ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_features=[feature for feature in numerical_features if len(train_data[feature].unique())<25 and feature not in year_features+['Id']]
discrete_features
```

```python
train_data[discrete_features].head()
```

```python
for feature in discrete_features:
    data=train_data.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_variable=[feature for feature in numerical_features if feature not in year_features+discrete_features+['Id']]
continuous_variable
```

```python
for feature in continuous_variable:
    data=train_data.copy()
    data[feature].plot(kind='hist',bins=25)
    plt.ylabel('Count')
    plt.xlabel(feature)
    plt.title(feature)
    plt.show()
```

```python
# Logarithmic Transformation
for feature in continuous_variable:
    data=train_data.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.title(feature)
        plt.show()
```

### Outliers

```python
for feature in continuous_variable:
    data=train_data.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
