```python
dataset1=dataset.copy()
dataset1['lat']=dataset1['lat'].astype('int')
dataset1.groupby('lat').count()
```

```python
dataset1.groupby('zip').count()
```

```python
dataset1['timeStamp']=dataset1['timeStamp'].astype('datetime64[ns]')
dataset1['Day']=dataset1['timeStamp'].dt.weekday_name
dataset1.head()
```

```python
def check(x):
    if str(x)[0].lower()=='a':
        return True
    else:
        return False
```

```python
dataset1[dataset1['addr'].apply(check)]['zip']
```
