```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

### Univariate Analysis

```python
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa),'r')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica),'b')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor),'y')
plt.xlabel('sepal length')
plt.show()
```

### Bivariate Analysis

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,'sepal_length','sepal_width')
plt.legend()
```

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,'petal_length','petal_width')
plt.legend()
```

### Multivariate Analysis

```python
sns.pairplot(df,hue='species')
```
